# glove vec model

```python
f = open('glove.6B.50d.txt')
```

```python
embeding_index = {}
cnt = 0

for line in f:
    
    values = line.split()
    word = values[0]
    coeff = np.asarray(values[1:], dtype = 'float')
    
    embeding_index[word] = coeff

f.close()
```

```python
embeding_index["was"]
```

```python
emb_dim = embeding_index["was"].shape[0]
print(emb_dim)
```

```python
def embedding_output(X):
    
    maxlen = 2500
    
    embedding_out = np.zeros((X.shape[0],maxlen,emb_dim))
    
    for ix in range(X.shape[0]):
        words = X[ix].split()
        
        for jx in range(len(words)):
            
            try:
                embedding_out[ix][jx] = embeding_index[words[jx].lower()]
            except:
                embedding_out[ix][jx] = np.zeros((50,))
            
    return embedding_out
```

```python
emb_train = np.zeros((0,2500,50))

for beg in range(0,1000,100):
    
    end = beg+100
    batch = X_train[beg:end]
    emb = embedding_output(batch)
    emb_train = np.append(emb_train,emb,axis=0)
```

```python
emb_test = np.zeros((0,2500,50))

for beg in range(0,500,100):
    
    end = beg+100
    batch = X_test[beg:end]
    emb = embedding_output(batch)
    emb_test = np.append(emb_test,emb,axis=0)
```

```python
print(emb_train.shape)
print(emb_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Activation, Dropout
from keras.utils import to_categorical
```

```python
model = Sequential()
model.add(LSTM(64, input_shape = (2500,50), return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('sigmoid'))
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
y_train = y_train[:1000]
y_test = y_test[:500]

y_train = to_categorical(y_train,num_classes=2)
y_test = to_categorical(y_test,num_classes=2)

print(y_train.shape)
print(y_test.shape)
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint("best_model.h5", monitor='val_loss', verbose=True, save_best_only=True)
earlystop = EarlyStopping(monitor='val_acc', patience=10)

hist = model.fit(emb_train,y_train,epochs=100,batch_size=64,shuffle=True,validation_split=0.2,callbacks=[checkpoint, earlystop])
```

```python
model.load_weights('best_model.h5')
```

```python
model.evaluate(emb_test,y_test)
```

# our own word_embedding

```python
def vect_sent(sentences, dims=10000):
    
    outputs = np.zeros((len(sentences),dims))
    
    for i,idx in enumerate(sentences):
        outputs[i,idx] = 1
    
    return outputs
```

```python
X_train = vect_sent(X_traindata)
X_test = vect_sent(X_testdata)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(X_train[0])
```

```python
y_train = np.asarray(y_traindata).astype('float32')
y_test = np.asarray(y_testdata).astype('float32')
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Activation, Dropout
from keras.utils import to_categorical
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape = (10000,)))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
X_val = X_train[:5000]
X_train_new = X_train[5000:]

y_val = y_train[:5000]
y_train_new = y_train[5000:]
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint("best_model.h5", monitor='val_loss', verbose=True, save_best_only=True)
earlystop = EarlyStopping(monitor='val_acc', patience=10)

hist = model.fit(X_train_new,y_train_new,epochs=20,batch_size=128,shuffle=True,validation_data=(X_val,y_val),callbacks=[checkpoint, earlystop])
```

```python
model.load_weights('best_model.h5')
model.evaluate(X_test,y_test)
```
